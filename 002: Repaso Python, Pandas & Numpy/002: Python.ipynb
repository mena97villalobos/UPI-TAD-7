# 5. Loops

```python
# Ejemplo de un FOR LOOP
for name in names:
    print(name)

    if name == 'buzz aldrin':
        print('Sup buzz \n')
        # Continuar al siguiente valor del FOR LOOP
        continue

    elif name == 'yuri gagarin':
        print('Hey yuri!')
        # TERMINAR/SALIRSE del FOR LOOP
        break

    # Caso por defecto
    else:
        # NO HACER NADA UN NO-OP
        pass
    print('Pleased to have you here!\n')
```

```python
# Hacer un ciclo por x cantidad de veces
# Recordar que range excluye el ultimo valor!!!!
for idx in range(1, 11, 2):
    print(idx)
```

```python
# Crear la variable launched
launched_shuttles = 0
total_shuttles = 10

# Crear un WHILE LOOP
while launched_shuttles <= total_shuttles:
    # Formatear un string utilizando {} y la función format()
    print('Shuttle {} launched'.format(launched_shuttles+1))
    launched_shuttles += 1

    if launched_shuttles == total_shuttles:
        print('All shuttles launched')
        # Condición de parada!!!!! Importante para no quedarnos enciclados
        break
```

# 5. Classes

```python
class Person():
    # Método que inicia una instacia de una clase
    def __init__(self, name, age, color):
        # Crear atributos para una persona
        self.name = name
        self.age = age
        self.color = color

    # Caluclar el año de nacimiento
    def year_of_birth(self):
        return 2024-self.age

    # Cualcular edad futura
    def project_age(self, years=5):
        return self.age+years
```

```python
new_person = Person('Elon Musk', 38, 'blue')
```

```python
# Acceder a un atributo de una clase
new_person.age
```

```python
# Correr un método de una clase
new_person.year_of_birth()
```

```python
# Correr un método con un argumento con nombre
new_person.project_age(years=10)
```

# 6. Herencia
- PARENT es la clase que le pasa atributos a sus hijos (los que heredan) = PERSON
- CHILD la clase que hereda atributos del padre = ASTRONAUT

```python
# Crear una clase CHILD
class Astronaut(Person):

    # Definir la función de inicialización
    def __init__(self, name, age, color, mission_length_in_months):
        # Antes que nada tenemos que llamar a nuestro padre!!!!
        super().__init__(name, age, color)
        self.mission_length_months = mission_length_in_months

    # Calcular la edad al retornar a la tierra
    def age_on_return(self):
        return self.project_age(years=int(self.mission_length_months/12))
```

```python
new_astronaut = Astronaut('Nick', 99, 'purple', 48)
```

```python
# Podemos acceder a los atributos de nuestra clase PARENT
new_astronaut.color
```

```python
# Al igual que a los métodos
new_astronaut.year_of_birth()
```

```python
# Run a method
new_astronaut.age_on_return()
```
